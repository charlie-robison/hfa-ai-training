- Company policies (return, warranty, shipping, remote work, PTO)

- Product descriptions

Each document is a dictionary with:
- `title` — The document name (used for citation/source tracking)
- `content` — The full text of the document
- `metadata` — Extra info like category, last_updated, department

1. Run this notebook first and use `%store` / `%store -r` to share the `DOCUMENTS` variable, **or**
2. Copy the `DOCUMENTS` list directly into the other notebooks.
