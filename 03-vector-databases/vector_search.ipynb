# Vector Search with Pinecone + OpenAI Embeddings

Embed grocery product documents, store them in Pinecone, and run semantic search.

```python
import os, json, time
from dotenv import load_dotenv
from pinecone import Pinecone, ServerlessSpec
from openai import OpenAI

load_dotenv(dotenv_path=os.path.join(os.path.dirname(os.getcwd()), ".env"))

pc = Pinecone(api_key=os.environ["PINECONE_API_KEY"])
openai = OpenAI(api_key=os.environ["OPENAI_API_KEY"])

EMBED_MODEL = "text-embedding-3-small"
EMBED_DIM   = 1536

def embed(texts: list[str]) -> list[list[float]]:
    res = openai.embeddings.create(model=EMBED_MODEL, input=texts)
    return [d.embedding for d in res.data]
```

```python
# --- Create index and upsert all products ---

INDEX = "grocery-products"

if INDEX in [i.name for i in pc.list_indexes()]:
    pc.delete_index(INDEX)

pc.create_index(
    name=INDEX,
    dimension=EMBED_DIM,
    metric="cosine",
    spec=ServerlessSpec(cloud="aws", region="us-east-1"),
)
time.sleep(1)
index = pc.Index(INDEX)

# Embed all product descriptions in one batch
texts = [p["text"] for p in products]
embeddings = embed(texts)

# Upsert with metadata (store text in metadata so we can retrieve it)
vectors = [
    {
        "id": p["id"],
        "values": emb,
        "metadata": {
            "name": p["name"],
            "category": p["category"],
            "price": p["price"],
            "organic": p["organic"],
            "text": p["text"],
        },
    }
    for p, emb in zip(products, embeddings)
]

index.upsert(vectors=vectors)
time.sleep(2)

print(f"Indexed {index.describe_index_stats().total_vector_count} products")
```

```python
# --- Semantic search helper ---

def search(query: str, top_k: int = 3, filter: dict = None):
    """Embed a query and search Pinecone. Print results."""
    vec = embed([query])[0]
    params = {"vector": vec, "top_k": top_k, "include_metadata": True}
    if filter:
        params["filter"] = filter

    results = index.query(**params)

    print(f'\nQuery: "{query}"')
    if filter:
        print(f"Filter: {filter}")
    print("-" * 60)
    for m in results.matches:
        print(f"  {m.score:.3f}  {m.metadata['name']}")
        print(f"         ${m.metadata['price']:.2f} | {m.metadata['category']}")
    print()
```

```python
# --- Queries: semantic meaning, not keyword matching ---

# "protein" isn't in every result, but the model understands the concept
search("high protein low fat option for working out")

# Finds sourdough + GF bread — understands "digestive" meaning
search("something easy on the stomach for sensitive digestion")

# Matches plant-based + organic products across categories
search("healthy vegan options")
```

```python
# --- Queries with metadata filters ---

# Only organic products
search("best quality natural products", filter={"organic": True})

# Only bread, under $5
search("everyday sandwich bread", filter={
    "$and": [{"category": "bread"}, {"price": {"$lt": 5.00}}]
})

# Eggs only
search("heart healthy breakfast", filter={"category": "eggs"})
```
